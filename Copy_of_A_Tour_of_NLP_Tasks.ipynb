### Pipelines

Easy way to use models for inference

Objects that abstract most of the complex code from the library, offering a simple API dedicated to several tasks such as:

(i) Named Entity Recognition

(ii) Masked Language Modeling

(iii) Sentiment Analysis

(iv) Summarization

(v) Question Answering

### Text Classification - Sentiment Analysis

Classify a given text into positive negative or neutral

```python
#hide_output
from transformers import pipeline

classifier = pipeline("text-classification")
```

```python
import pandas as pd

outputs = classifier(text)
pd.DataFrame(outputs)
```

```python
text1 = "I love huggingface"
# change the text and check
outputs = classifier(text1)
pd.DataFrame(outputs)
```

### Named Entity Recognition

In a given text indentify name of person, location, things etc

```python
ner_tagger = pipeline("ner", aggregation_strategy="simple")
outputs = ner_tagger(text)
pd.DataFrame(outputs)
```

### Question Answering

Given a question and context, this pipeline finds the answer for the question in the context

```python
reader = pipeline("question-answering")
# change the question and check
question = "Why is the customer happy?"
outputs = reader(question=question, context=text)
pd.DataFrame([outputs])
```

### Summarization

Summarize a given text. Given a text make a shorter version of it without losing meaning

```python
# you may change the max_length and check
summarizer = pipeline("summarization")
outputs = summarizer(text, max_length=100, clean_up_tokenization_spaces=True)
print(outputs[0]['summary_text'])
```

### Translation

Convert a text in one language to another

```python
# create a pipeline to convert from English to German
translator = pipeline("translation_en_to_de",
                      model="Helsinki-NLP/opus-mt-en-de")
outputs = translator(text, clean_up_tokenization_spaces=True, min_length=100)
print(outputs[0]['translation_text'])
```

### Text Generation

Understand a text T and generate text related to the original text T


```python
#hide
from transformers import set_seed
set_seed(42) # Set the seed to get reproducible results
```

```python
generator = pipeline("text-generation")
response = "Dear Bumblebee, I am sorry to hear that your order was mixed up."
prompt = text + "\n\nCustomer service response:\n" + response
outputs = generator(prompt, max_length=200)
print(outputs[0]['generated_text'])
```
